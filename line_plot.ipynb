```python
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
import numpy as np
```

```python
x = [10, 20, 30, 40, 50, 60, 70]
y = [13, 16, 10, 35, 28, 35, 50]

plt.plot(x, y, color='orange', linestyle='-', linewidth=3)
plt.show()
```

```python
x = [1, 2, 3]
y = np.array([[1, 2], [3, 4], [5, 6]])

plt.plot(x, y)
plt.show()

# equivalent to
for col in range(y.shape[1]):
    plt.plot(x, y[:, col])
```

### Multiplots

```python
# subplot(nrows, ncols, index, **kwargs)
plt.subplot(2, 2, 1)
plt.plot([1, 2, 3, 4], [1, 2, 3, 4])
plt.title('A', color='red')

plt.subplot(2, 2, 2)
plt.plot([1, 2, 3, 4], [1, 2, 3, 4])
plt.title('B', color='red')

plt.subplot(2, 2, 3)
plt.plot([1, 2, 3, 4], [1, 2, 3, 4])
plt.title('C', color='red')

plt.subplot(2, 2, 4)
plt.plot([1, 2, 3, 4], [1, 2, 3, 4])
plt.title('D', color='red')
plt.show()
```

```python
fig4 = plt.figure(constrained_layout=True)
spec4 = fig4.add_gridspec(ncols=2, nrows=2)
anno_opts = dict(xy=(0.5, 0.5), xycoords='axes fraction', va='center', ha='center')

fig4.add_subplot(spec4[0, 0]).annotate('GridSpec[0, 0]', **anno_opts)
fig4.add_subplot(spec4[0, 1]).annotate('GridSpec[0, 1]', **anno_opts)
fig4.add_subplot(spec4[1, 0]).annotate('GridSpec[1, 0]', **anno_opts)
fig4.add_subplot(spec4[1, 1]).annotate('GridSpec[1, 1]', **anno_opts)
```

```python
x = np.arange(1, 5)
fig, a = plt.subplots(2, 2)

a[0][0].plot(x, x*x)
a[0][0].set_title('Square', color='g')

a[0][1].plot(x, np.sqrt(x))
a[0][1].set_title('Square root', color='g')

a[1][0].plot(x, np.exp(x))
a[1][0].set_title('exp', color='g')

a[1][1].plot(x, np.log10(x))
a[1][1].set_title('log', color='g')
```
