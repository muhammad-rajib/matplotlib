```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# using object
x = np.arange(1, 11)
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**3, 'g', lw=2)
axes[0].set_title('Default Grid')
axes[0].grid(True)

axes[1].plot(x, np.exp(x), 'r', lw=2)
axes[1].set_title('Custom Grid')
axes[1].grid(color='b', ls='-.', lw=0.5)

axes[2].plot(x, x, lw=2)
axes[2].set_title('No Grid')

fig.tight_layout()
plt.show()
```

```python
# using function system
x = np.arange(1, 11)

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.plot(x, x**2, 'g', lw=2)
plt.title('Default Grid')
plt.grid()

plt.subplot(1, 3, 2)
plt.plot(x, np.exp(x), 'r', lw=2)
plt.title('Custom Grid')
plt.grid(color='b', ls='-.')

plt.subplot(1, 3, 3)
plt.plot(x, x, lw=2)
plt.title('No Grid')

plt.tight_layout()
plt.show()
```
