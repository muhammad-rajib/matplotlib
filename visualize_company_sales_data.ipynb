### Visualize Company Sales Data using Matplotlib

```python
# import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
# load company sales data
sales_data = pd.read_csv('company_sales_data.csv')
sales_data
```

### Exercise 1: Read Total profit of all months and show it using a line plot

```python
total_profit = sales_data['total_profit']
month_number = sales_data['month_number']

plt.plot(month_number, total_profit, marker='D', linewidth=2)

plt.title('Company profit per month')
plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.xticks(month_number)
plt.yticks([100000, 200000, 300000, 400000, 500000])

plt.show()
```

### Exercise 2: Get total profit of all months and show line plot with the following Style properties

```python
total_profit = sales_data['total_profit']
month_list = sales_data['month_number']

plt.plot(month_list, total_profit, color='red', marker='o', linewidth=3, 
         markerfacecolor='k', linestyle='--', label='Profit data of last year')

plt.title('Company sales data of last year')
plt.xlabel('Month Number')
plt.ylabel('Solid Units Number')
plt.xticks(month_list)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.legend(loc='lower right', shadow=True)

plt.show()
```

### Exercise 3: Read all product sales data and show it  using a multiline plot

```python
# sales data of company
facecream = sales_data['facecream']
facewash = sales_data['facewash']
toothpaste = sales_data['toothpaste']
bathingsoap = sales_data['bathingsoap']
shampoo = sales_data['shampoo']
moisturizer = sales_data['moisturizer']

# plot chart
plt.plot(facecream, marker='o', color='red', label='Face cream Sales Data')
plt.plot(facewash, marker='o', color='orange', label='Face Wash Sales Data')
plt.plot(toothpaste, marker='o', color='green', label='Tooth Paste Sales Data')
plt.plot(bathingsoap, marker='o', color='brown', label='Bathing Soap Sales Data')
plt.plot(shampoo, marker='o', color='gold', label='Shampoo Sales Data')
plt.plot(moisturizer, marker='o', color='olive', label='Moisturizer Sales Data')

# xlabel, ylabel, xticks, yticks
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.xticks(sales_data['month_number'])
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])

plt.title('Sales Data')
plt.legend(loc='upper left', shadow=True)

plt.show()
```

### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot

```python
number_of_units = np.arange(4500, 8500, step=500)
month_number = sales_data['month_number']
toothpaste = sales_data['toothpaste']

plt.scatter(month_number, toothpaste, label='Tooth paste Sales data')
plt.title('Tooth paste Sales data')
plt.xlabel('Month Number')
plt.ylabel('Number units Sold')
plt.xticks(month_number)
plt.yticks(number_of_units)
plt.grid(linewidth=1, linestyle='--')
plt.legend(loc='upper left', shadow=True)

plt.show()
```

### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

```python
# sales data
facecream = sales_data['facecream']
facewash = sales_data['facewash']
month_list = sales_data['month_number']
y_units = np.arange(0, 4000, step=500).tolist()

# visualization
plt.bar(month_list-0.25, facecream, width=0.25, label='Face Cream sales data', align='edge')
plt.bar(month_list+0.25, facewash, width=-0.25, label='Face Wash sales data', align='edge')

plt.title('Facewash and facecream slaes data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.xticks(month_list)
plt.yticks(y_units)
plt.legend(loc='upper left')
plt.grid(True, linewidth=1, linestyle='--')

plt.show()
```
