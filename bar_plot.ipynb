```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
langs = ['C', 'C++', 'Java', 'PHP', 'Python']
students = [23, 17, 30, 20, 33]

plt.bar(langs, students)
plt.show()
```

```python
data = [[30, 25, 50, 20], [40, 23, 51, 17], [35, 22, 45, 19]]
x = np.arange(4)

plt.bar(x, data[0], color='b', width=0.25, label='CS')
plt.bar(x+0.25, data[1], color='g', width=0.25, label='IT')
plt.bar(x+0.50, data[2], color='r', width=0.25, label='E&TC')
plt.xticks(x+0.25, [2015, 2016, 2017, 2018])

plt.legend(shadow=True)
plt.show()
```
