```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
x = np.arange(1, 8)
y = [50, 51, 52, 48, 47, 49, 46]

# visualization 
plt.plot(x, y, color='g', linewidth=3, linestyle='dotted')
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
```

### Axes labels, Legend, Grid

```python
days = [1, 2, 3, 4, 5, 6, 7]
max_temp = [50, 51, 52, 48, 47, 50, 46]
min_temp = [43, 40, 45, 38, 44, 35, 37]
avg_temp = [45, 48, 46, 49, 44, 45, 43]

plt.title('Weather')
plt.xlabel('Days')
plt.ylabel('Temperature')

plt.plot(days, max_temp, label='Max')
plt.plot(days, min_temp, label='Min')
plt.plot(days, avg_temp, label='Average')
plt.legend(loc='best', shadow=True)

# add grid
plt.grid()
```

### Bar Chart

```python
company = ['Google', 'Amazon', 'Microsoft', 'Facebook']
revenue = [90, 136, 89, 37]
profit = [45, 80, 35, 25]

xpos = np.arange(len(company))
plt.bar(xpos-0.2, revenue, width=0.4, label='Revenue')
plt.bar(xpos+0.2, profit, width=0.4, label='Profit')

plt.title('US Tech Stocks')
plt.xticks(ypos, company)
plt.ylabel('revenue(bln)')
plt.legend(loc='best', shadow=True)
```

```python
xpos = np.arange(len(company))
plt.barh(xpos-0.2, revenue, left=1, label='Revenue')
plt.barh(xpos+0.2, profit, label='Profit')

plt.title('US Tech Stocks')
plt.yticks(ypos, company)
plt.xlabel('revenue(bln)')
plt.legend(loc='best', shadow=True)
```

## Histograms

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 135, 150, 84, 69, 89, 79, 120, 112, 100]

plt.title('Patient Blood Sugar Info')
plt.xlabel('Blood Sugar Range')
plt.ylabel('Total number of patient')
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100, 125, 150], rwidth=0.90, 
         color=['b', 'orange'], label=['men', 'women'])
plt.legend(shadow=True)
```

```python
# Horizontal Histogram
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 135, 150, 84, 69, 89, 79, 120, 112, 100]

plt.title('Patient Blood Sugar Info')
plt.xlabel('Blood Sugar Range')
plt.ylabel('Total number of patient')
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100, 125, 150], rwidth=0.90, 
         color=['b', 'orange'], label=['men', 'women'], orientation='horizontal')
plt.legend(shadow=True)
```

## Pie Chart

```python
exp_labels = ['Home Rent', 'Food', 'Phone/Internet Bill', 'Car', 'Other Utilities']
exp_vals = [1400, 600, 300, 410, 250]

plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.1f%%',
       explode=[0, 0, 0, 0, 0.2], startangle=30)

# SAVE FILE AS A PNG
plt.savefig('piechart.png', bbox_inches='tight', pad_inches=2, transparent=True)

# SAVE FILE AS A PDF
plt.savefig('piechart.pdf', bbox_inches='tight', pad_inches=2, transparent=True)
```
